```python
df = pd.read_csv('healthcare_dataset.csv')
```

```python
print("Displaying the first few rows of the DataFrame:")
print(df.head())
```

```python
print("Null values before replacement:")
print(df.isnull().sum())
```

```python
df.dropna(inplace=True)
```

```python
print("Null values after removing:")
print(df.isnull().sum())
```

```python
from sklearn.preprocessing import MinMaxScaler
billing_amount = df[['Billing Amount']]
scaler = MinMaxScaler()
normalized_billing_amount = scaler.fit_transform(billing_amount)
df['Normalized Billing Amount'] = normalized_billing_amount
print(df.head())
df.to_csv('/content/normalized_healthcare_dataset.csv', index=False)
```

```python
df['Billing Amount'] = df['Billing Amount'].apply(np.int64)
print(df.head())
```

```python
print(df.head())
def mc(condition):
    if condition in ['Cancer', 'Leukemia', 'Lymphoma']:
        return 'Severe'
    elif condition in ['Diabetes', 'Hypertension', 'Obesity']:
        return 'Chronic'
    elif condition in ['Cold', 'Flu', 'Allergy']:
        return 'Minor'
    else:
        return 'Other'

df['Medical Condition Discretized'] = df['Medical Condition'].apply(mc)
print(df.head())
df.to_csv('/content/discretized_healthcare_dataset.csv', index=False)
```
